Para las funciones de transferencia en lazo abierto encontradas en el punto anterior, encuentre la expresion de la salida en el tiempo para una configuracion en lazo cerrado del sistema, en funcion de R, L y C, para las siguientes entradas:i) impulso, ii) escalon unitario, y iv) rampa. Compruebe los resultados en simulacion sobre Python (utililzando sympy) y grafique la senal de salida para cada caso.


Ahora realizaremos la simulación numérica y graficaremos las salidas para valores de ejemplo de R, L y C. Puedes cambiar estos valores para observar cómo afecta la respuesta del sistema.

```python
# Realizar simulaciones numéricas y graficar las salidas

# Define valores de ejemplo para R, L, C
R_val = 10
L_val = 1
C_val = 0.1

# Crea un rango de tiempo
tiempo = np.linspace(0, 10, 500)

# Sustituir valores numéricos y convertir las expresiones simbólicas a funciones numéricas
# RLC Serie
get_y_rlc_serie_impulso = sp.lambdify(t, y_rlc_serie_impulso.subs({R: R_val, L: L_val, C: C_val}), 'numpy')
get_y_rlc_serie_escalon = sp.lambdify(t, y_rlc_serie_escalon.subs({R: R_val, L: L_val, C: C_val}), 'numpy')
get_y_rlc_serie_rampa = sp.lambdify(t, y_rlc_serie_rampa.subs({R: R_val, L: L_val, C: C_val}), 'numpy')

# RCL Paralelo
get_y_rcl_paralelo_impulso = sp.lambdify(t, y_rcl_paralelo_impulso.subs({R: R_val, L: L_val, C: C_val}), 'numpy')
get_y_rcl_paralelo_escalon = sp.lambdify(t, y_rcl_paralelo_escalon.subs({R: R_val, L: L_val, C: C_val}), 'numpy')
get_y_rcl_paralelo_rampa = sp.lambdify(t, y_rcl_paralelo_rampa.subs({R: R_val, L: L_val, C: C_val}), 'numpy')


# Calcular los valores de la salida en el tiempo
# RLC Serie
salida_rlc_serie_impulso = get_y_rlc_serie_impulso(tiempo)
salida_rlc_serie_escalon = get_y_rlc_serie_escalon(tiempo)
salida_rlc_serie_rampa = get_y_rlc_serie_rampa(tiempo)

# RCL Paralelo
salida_rcl_paralelo_impulso = get_y_rcl_paralelo_impulso(tiempo)
salida_rcl_paralelo_escalon = get_y_rcl_paralelo_escalon(tiempo)
salida_rcl_paralelo_rampa = get_y_rcl_paralelo_rampa(tiempo)


# Graficar los resultados para el circuito RLC Serie
plt.figure(figsize=(12, 8))
plt.suptitle('Simulación Circuito RLC Serie en Lazo Cerrado', fontsize=16)

plt.subplot(3, 1, 1)
plt.plot(tiempo, salida_rlc_serie_impulso)
plt.title('Respuesta al Impulso (R={}, L={}, C={})'.format(R_val, L_val, C_val))
plt.xlabel('Tiempo (t)')
plt.ylabel('Amplitud')
plt.grid(True)

plt.subplot(3, 1, 2)
plt.plot(tiempo, salida_rlc_serie_escalon)
plt.title('Respuesta al Escalón Unitario (R={}, L={}, C={})'.format(R_val, L_val, C_val))
plt.xlabel('Tiempo (t)')
plt.ylabel('Amplitud')
plt.grid(True)

plt.subplot(3, 1, 3)
plt.plot(tiempo, salida_rlc_serie_rampa)
plt.title('Respuesta a la Rampa (R={}, L={}, C={})'.format(R_val, L_val, C_val))
plt.xlabel('Tiempo (t)')
plt.ylabel('Amplitud')
plt.grid(True)

plt.tight_layout(rect=[0, 0.03, 1, 0.95]) # Ajustar layout para el título
plt.show()

# Graficar los resultados para el circuito RCL Paralelo
plt.figure(figsize=(12, 8))
plt.suptitle('Simulación Circuito RCL Paralelo en Lazo Cerrado', fontsize=16)

plt.subplot(3, 1, 1)
plt.plot(tiempo, salida_rcl_paralelo_impulso)
plt.title('Respuesta al Impulso (R={}, L={}, C={})'.format(R_val, L_val, C_val))
plt.xlabel('Tiempo (t)')
plt.ylabel('Amplitud')
plt.grid(True)

plt.subplot(3, 1, 2)
plt.plot(tiempo, salida_rcl_paralelo_escalon)
plt.title('Respuesta al Escalón Unitario (R={}, L={}, C={})'.format(R_val, L_val, C_val))
plt.xlabel('Tiempo (t)')
plt.ylabel('Amplitud')
plt.grid(True)

plt.subplot(3, 1, 3)
plt.plot(tiempo, salida_rcl_paralelo_rampa)
plt.title('Respuesta a la Rampa (R={}, L={}, C={})'.format(R_val, L_val, C_val))
plt.xlabel('Tiempo (t)')
plt.ylabel('Amplitud')
plt.grid(True)

plt.tight_layout(rect=[0, 0.03, 1, 0.95]) # Ajustar layout para el título
plt.show()
```
