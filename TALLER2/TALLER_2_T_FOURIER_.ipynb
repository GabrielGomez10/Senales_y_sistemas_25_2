# $ Series de Fourier (Exponencial, Trigonométrica y Compacta)$

#$Aplicación:$
 Se usan para representar señales periódicas en el tiempo continuo.
Espectro: Tienen un espectro de frecuencia discreto. Esto significa que la señal está compuesta por una suma de senoides a frecuencias específicas (múltiplos de la frecuencia fundamental).
Similitudes entre las formas (Exponencial, Trigonométrica y Compacta): Son representaciones equivalentes de la misma serie, solo cambian la forma en que se expresan los coeficientes y las bases (senos/cosenos vs exponenciales complejas).
Transformada de Fourier

#Aplicación:
 Se usa para representar señales no periódicas en el tiempo continuo.
Espectro: Tienen un espectro de frecuencia continuo. Esto significa que la señal puede contener componentes en cualquier frecuencia.
Transformada de Fourier en Tiempo Discreto (DTFT)

Aplicación: Se usa para representar señales no periódicas en el tiempo discreto.
Espectro: Tienen un espectro de frecuencia continuo (aunque periódico en la frecuencia).
Transformada Discreta de Fourier (DFT)

Aplicación: Se usa para representar señales periódicas en el tiempo discreto. En la práctica, se usa para analizar señales de tiempo discreto de longitud finita (considerándolas como un período de una señal periódica).
Espectro: Tienen un espectro de frecuencia discreto.
Relación con la DTFT: La DFT son muestras del espectro continuo de la DTFT en frecuencias específicas.
Combinaciones de Espectro y Señal

Señal / Espectro	Continuo	Discreto
Tiempo Continuo	Transformada Fourier	Serie de Fourier
Tiempo Discreto	DTFT	DFT
Algoritmo Fast Fourier Transform (FFT)

¿Qué es? La FFT no es una transformada nueva, sino un algoritmo eficiente para calcular la Transformada Discreta de Fourier (DFT).
Utilidad: La utilidad principal de la FFT es reducir drásticamente el costo computacional para calcular la DFT.
Funcionamiento (En detalle): La FFT se basa en la propiedad de "dividir y conquistar". Descompone el cálculo de la DFT de tamaño N en cálculos de DFTs de tamaño menor (por ejemplo, N/2), y luego combina los resultados. Esto es posible si N es un número compuesto (especialmente si es una potencia de 2). El algoritmo más común es el de Cooley-Tukey. Básicamente, agrupa los términos de la suma de la DFT de una manera inteligente para evitar repetir cálculos.
Costo Computacional:
DFT directa: Calcular la DFT directamente requiere aproximadamente O(N²) operaciones (multiplicaciones y sumas), donde N es el número de muestras.
FFT: El algoritmo FFT (para N que es potencia de 2) requiere aproximadamente O(N log₂ N) operaciones.
Comparación de Costo: Como puedes ver, O(N log₂ N) crece mucho más lento que O(N²). Para valores grandes de N, la FFT es significativamente más rápida que el cálculo directo de la DFT. Por ejemplo, para N = 1024, N² es aproximadamente un millón, mientras que N log₂ N es aproximadamente 10000, una reducción muy considerable.

#$La modulación por amplitud (AM)$
La modulación por amplitud (AM)  con detección coherente es un método de demodulación que utiliza una portadora local sincronizada en frecuencia y fase con la portadora original utilizada en la modulación. Esto permite recuperar la señal mensaje de manera más fiel en comparación con la detección de envolvente, especialmente cuando el índice de modulación es alto.

¿En qué consiste la detección coherente?

Multiplicación: La señal AM recibida se multiplica por una señal sinusoidal local que tiene la misma frecuencia ($\omega_c$$\omega_c$) y fase ($\phi$$\phi$) que la portadora original. La señal recibida AM es de la forma $s(t) = A_c(1 + m k_a f(t)) \cos(\omega_c t)$$s(t) = A_c(1 + m k_a f(t)) \cos(\omega_c t)$, donde $A_c$$A_c$ es la amplitud de la portadora, $m$$m$ es el índice de modulación, $k_a$$k_a$ es una constante y $f(t)$$f(t)$ es la señal mensaje. La señal local es $2 \cos(\omega_c t + \phi)$$2 \cos(\omega_c t + \phi)$. El factor de 2 se usa a menudo para simplificar el resultado final.
Filtrado Pasa Bajo (LPF): El resultado de la multiplicación contiene componentes de alta frecuencia (alrededor de $2\omega_c$$2\omega_c$) y componentes de baja frecuencia. Un filtro pasa bajo se utiliza para eliminar las componentes de alta frecuencia, dejando solo las componentes de baja frecuencia que son proporcionales a la señal mensaje.
Aplicaciones:

La detección coherente en AM se utiliza en situaciones donde se requiere una demodulación más robusta y fiel, a menudo en sistemas de comunicación donde la relación señal-ruido puede ser baja. Algunas aplicaciones incluyen:

Sistemas de comunicación militares y de seguridad: Donde la fiabilidad de la comunicación es crucial.
Algunos sistemas de radioaficionados: Para obtener una mejor calidad de audio.
Receptores de radio de alta fidelidad: Aunque la detección de envolvente es más común en receptores comerciales debido a su simplicidad.
Comunicaciones digitales (como ASK - Amplitude Shift Keying): Donde la detección coherente es esencial para determinar el estado de la señal (bit 0 o 1).

#$EJEMPLO$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import rfft, rfftfreq

def plot_signal_time_freq(t, signal, title, freq_range=None):
    """Plots the signal in time and frequency domains."""
    plt.figure(figsize=(12, 6))

    # grafico en el domino del tiempo
    plt.subplot(2, 1, 1)
    plt.plot(t, signal)
    plt.title(f'{title} - Dominio del Tiempo')
    plt.xlabel('Tiempo [s]')
    plt.ylabel('Amplitud')
    plt.grid(True)

    # grafico en el domino de la frecuencia
    plt.subplot(2, 1, 2)
    n = len(signal)
    yf = rfft(signal)
    xf = rfftfreq(n, t[1] - t[0])

    # establecer el rango de frecuencia especifico
    plt.plot(xf, np.abs(yf))
    plt.title(f'{title} - Dominio de la Frecuencia (Magnitud)')
    plt.xlabel('Frecuencia [Hz]')
    plt.ylabel('Magnitud')
    plt.grid(True)

    # Set frequency range if specified
    if freq_range:
        plt.xlim(freq_range)

    plt.tight_layout()
    plt.show()

# --- Parámetros de simulación ---
fs = 10000  # Frecuencia de muestreo
T = 1 / fs
t = np.arange(0, 1, T) # Vector de tiempo
fc = 500  # Frecuencia de la portadora
Ac = 1    # Amplitud de la portadora

# Índice de modulación (define el usuario)
m = 0.8 # Puedes cambiar este valor para ver el efecto

# --- Señal Portadora ---
carrier = Ac * np.cos(2 * np.pi * fc * t)

# --- Señal Mensaje 1: Pulso Rectangular ---
message_rect_duration = 0.2
message_rect_amplitude = 0.5
message_rect = np.zeros_like(t)
start_time_rect = 0.4
end_time_rect = start_time_rect + message_rect_duration
message_rect[(t >= start_time_rect) & (t <= end_time_rect)] = message_rect_amplitude

# Modulación AM para el pulso rectangular
am_rect = Ac * (1 + m * message_rect) * np.cos(2 * np.pi * fc * t)

# Detección Coherente para el pulso rectangular
# Señal portadora local (sincronizada idealmente)
local_carrier_rect = 2 * np.cos(2 * np.pi * fc * t)
demodulated_rect_raw = am_rect * local_carrier_rect

# Filtrado Pasa Bajo (ejemplo simple con un filtro promedio móvil)
# En una aplicación real, se usaría un filtro LPF diseñado adecuadamente.
window_size = 50 # Ajusta este valor según sea necesario
demodulated_rect_filtered = np.convolve(demodulated_rect_raw, np.ones(window_size)/window_size, mode='same')

# --- Señal Mensaje 2: Coseno ---
fm = 50  # Frecuencia de la señal mensaje coseno
message_cos = 0.5 * np.cos(2 * np.pi * fm * t) # Amplitud del coseno para evitar sobremodulación con m=0.8

# Modulación AM para el coseno
am_cos = Ac * (1 + m * message_cos) * np.cos(2 * np.pi * fc * t)

# Detección Coherente para el coseno
local_carrier_cos = 2 * np.cos(2 * np.pi * fc * t)
demodulated_cos_raw = am_cos * local_carrier_cos
demodulated_cos_filtered = np.convolve(demodulated_cos_raw, np.ones(window_size)/window_size, mode='same')

# --- Graficar señales ---

# Señal Mensaje Pulso Rectangular
plot_signal_time_freq(t, message_rect, 'Señal Mensaje - Pulso Rectangular')

# Señal Modulada AM (Pulso Rectangular)
plot_signal_time_freq(t, am_rect, 'Señal Modulada AM - Pulso Rectangular', freq_range=[0, fc * 2 + fm * 2])

# Señal Demodulada (después de multiplicar por la portadora local) - Pulso Rectangular
plot_signal_time_freq(t, demodulated_rect_raw, 'Señal Demodulada (Antes LPF) - Pulso Rectangular', freq_range=[0, fc * 2 + fm * 2])

# Señal Demodulada (después del LPF) - Pulso Rectangular
plot_signal_time_freq(t, demodulated_rect_filtered, 'Señal Demodulada (Después LPF) - Pulso Rectangular', freq_range=[0, fm * 2 + 10])


# Señal Mensaje Coseno
plot_signal_time_freq(t, message_cos, 'Señal Mensaje - Coseno')

# Señal Modulada AM (Coseno)
plot_signal_time_freq(t, am_cos, 'Señal Modulada AM - Coseno', freq_range=[0, fc * 2 + fm * 2])

# Señal Demodulada (después de multiplicar por la portadora local) - Coseno
plot_signal_time_freq(t, demodulated_cos_raw, 'Señal Demodulada (Antes LPF) - Coseno', freq_range=[0, fc * 2 + fm * 2])

# Señal Demodulada (después del LPF) - Coseno
plot_signal_time_freq(t, demodulated_cos_filtered, 'Señal Demodulada (Después LPF) - Coseno', freq_range=[0, fm * 2 + 10])
```

```python
!pip install yt-dlp pydub ffmpeg-python --quiet

import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, ifft, fftfreq
from pydub import AudioSegment
from IPython.display import Audio, display
import yt_dlp
import os

# ✅ PARTE 1: DESCARGAR Y RECORTAR AUDIO (SEG 20 A 25)
url = "https://www.youtube.com/watch?v=rkz6vcrRWOM"  # Cambia este enlace por tu canción favorita

ydl_opts = {
    'format': 'bestaudio/best',
    'outtmpl': 'audio.%(ext)s',
    'postprocessors': [{
        'key': 'FFmpegExtractAudio',
        'preferredcodec': 'mp3',
    }],
}

with yt_dlp.YoutubeDL(ydl_opts) as ydl:
    ydl.download([url])

# ✅ PARTE 2: CARGAR AUDIO Y CONVERTIR A SEÑAL
audio = AudioSegment.from_file("audio.mp3").set_channels(1).set_frame_rate(44100)
audio_fragment = audio[20000:25000]  # 5 segundos desde el segundo 20
Fs = audio_fragment.frame_rate
samples = np.array(audio_fragment.get_array_of_samples()).astype(np.float32) / 2**15
t = np.linspace(0, len(samples)/Fs, len(samples))

# ✅ PARTE 3: SEÑAL MENSAJE m(t)
m_t = samples

# ✅ PARTE 4: SEÑAL PORTADORA c(t)
Ac = 1.0
Fc = 10000  # 10 kHz
c_t = Ac * np.cos(2 * np.pi * Fc * t)

# ✅ PARTE 5: MODULACIÓN AM (índice de modulación = 1)
y_t = (1 + m_t) * c_t

# ✅ PARTE 6: FUNCIÓN FFT
def compute_fft(signal, Fs):
    N = len(signal)
    freqs = fftfreq(N, 1/Fs)
    spectrum = fft(signal)
    return freqs[:N//2], np.abs(spectrum[:N//2]) / N

# FFT señales
f_m, M_f = compute_fft(m_t, Fs)
f_c, C_f = compute_fft(c_t, Fs)
f_y, Y_f = compute_fft(y_t, Fs)

# ✅ PARTE 7: REPRODUCCIÓN DE AUDIO
print("🎧 Señal mensaje:")
display(Audio(m_t, rate=Fs))
print("🎧 Señal portadora:")
display(Audio(c_t, rate=Fs))
print("🎧 Señal modulada:")
display(Audio(y_t, rate=Fs))

# ✅ PARTE 8: GRÁFICAS EN TIEMPO
plt.figure(figsize=(15, 8))

plt.subplot(3, 1, 1)
plt.plot(t, m_t)
plt.title("Señal mensaje m(t)")
plt.grid()

plt.subplot(3, 1, 2)
plt.plot(t, c_t)
plt.title("Señal portadora c(t)")
plt.grid()

plt.subplot(3, 1, 3)
plt.plot(t, y_t)
plt.title("Señal modulada y(t)")
plt.grid()

plt.tight_layout()
plt.show()

# ✅ PARTE 9: GRÁFICAS EN FRECUENCIA
plt.figure(figsize=(15, 8))

plt.subplot(3, 1, 1)
plt.plot(f_m, M_f)
plt.title("Espectro |M(f)| de la señal mensaje")
plt.grid()

plt.subplot(3, 1, 2)
plt.plot(f_c, C_f)
plt.title("Espectro |C(f)| de la portadora")
plt.grid()

plt.subplot(3, 1, 3)
plt.plot(f_y, Y_f)
plt.title("Espectro |Y(f)| de la señal AM")
plt.grid()

plt.tight_layout()
plt.show()

# ============================
# PARTE 10: DEMODULACIÓN IDEAL POR FILTRADO FFT (PASA BAJAS)
# ============================

# Etapa 1: Mezcla (multiplicación de la señal modulada por la portadora)
mixed_signal = y_t * c_t  # Demodulación por multiplicación con portadora (receptor coherente)

# FFT de la señal mezclada (antes del filtro)
Mixed_fft = fft(mixed_signal)
frequencies = fftfreq(len(mixed_signal), 1/Fs)

# Filtro pasa bajas ideal (hasta 8kHz)
Mixed_fft_filtered = Mixed_fft.copy()
Mixed_fft_filtered[np.abs(frequencies) > 8000] = 0

# Señal demodulada después del filtro
demod_signal = np.real(ifft(Mixed_fft_filtered))

# Reproducción de audio señal demodulada
print("🎧 Señal demodulada:")
display(Audio(demod_signal, rate=Fs))

# ============================
# PARTE 11: GRÁFICOS DE LA ETAPA DE DEMODULACIÓN
# ============================

plt.figure(figsize=(15, 8))

plt.subplot(3, 1, 1)
plt.plot(t, mixed_signal)
plt.title("Señal mezclada (modulada × portadora) - antes del filtro pasa bajas")
plt.grid()

plt.subplot(3, 1, 2)
plt.plot(frequencies[:len(frequencies)//2], np.abs(Mixed_fft)[:len(frequencies)//2] / len(Mixed_fft))
plt.title("Espectro de la señal mezclada (antes del filtro)")
plt.grid()

plt.subplot(3, 1, 3)
plt.plot(t, demod_signal)
plt.title("Señal demodulada (después del filtro pasa bajas ideal)")
plt.grid()

plt.tight_layout()
plt.show()

# ============================
# PARTE 12: COMPARACIÓN VISUAL ENTRE m(t) y SEÑAL DEMODULADA
# ============================

plt.figure(figsize=(12, 5))
plt.plot(t, demod_signal, label="Señal demodulada")
plt.plot(t, m_t, alpha=0.5, label="Señal original m(t)")
plt.title("Comparación señal demodulada vs mensaje")
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()

# ============================
# LIMPIAR ARCHIVOS TEMPORALES
# ============================

os.remove("audio.mp3")
```

#$Aplicación \ en\ circuitos\ eléctricos - potencia.$

### Distorsión Total de Armónicos (THD)
La distorsión armónica total (THD o THDi) es una medida que indica cuánto se desvía una señal periódica de una forma de onda sinusoidal pura. Se expresa como el cociente entre el valor eficaz de todos los armónicos presentes (exceptuando el fundamental) y el valor eficaz de la componente fundamental:

$$THD = \dfrac{\sqrt{V^{2}_{2} + V^{2}_{3} + V^{2}_{4}+...}}{V_{1}}$$

Donde:

$I_{1}$ es el valor eficaz de la corriente fundamental

$I_{2}, I_{3}, ...$ son los valores eficaces de los armónicos superiores

La THD suele expresarse en porcentaje o en dB respecto a la fundamental como atenuación de la distorsión.

### Factor de Potencia
El factor de potencia (PF) es la relación entre la potencia real (activa) y la potencia aparente en un sistema eléctrico:

$$f.d.p = \dfrac{P}{S}$$

#### Distorsión del factor de potencia con base al THD

Cuando hay distorsión armónica (señales no sinusoidales), el factor de potencia se compone de dos factores:

- Factor de desplazamiento (DPF) : el coseno del ángulo de fase entre voltaje y corriente fundamental.
- Factor de distorsión (THD) : relacionado con la presencia de armónicos en la corriente.
Entonces:

$$f.d.p = DPF \cdot \dfrac{1}{\sqrt{1+THD^2}}$$

Esto nos permite estimar la distorsión del factor de potencia causada por los armónicos, si ya se conoce el THD.

#### Cálculo del THD desde la FFT
Para calcular el THD usando una Transformada Rápida de Fourier (FFT):

1. Se muestrea la señal de corriente (o voltaje).
2. Aplicamos la FFT para obtener las magnitudes de cada frecuencia.
3. Identificamos la magnitud del armónico fundamental ($I_{1}$).
4. Identificamos y sumamos las magnitudes de los armónicos superiores ($I_{2},I_{3}$)
5. Usamos la fórmula del THD.

### Ejemplo: Rectificador de Onda Completa
- ### Caso: Carga resistiva pura
  - Fuente AC: $V_o(t)=V_m \cdot sin(ωt)$

  - Frecuencia: $f=60Hz$

  - Carga: $R=100Ω $

  Suponiendo que la magnitud fundamental es de $I_1 = 1A$, y que tenemos 4 armónicos superiores:

  $I_2 = 0.2A$,

  $I_3 = 0.1A$,

  $I_4 = 0.05A$,

  $I_5 = 0.03A$

  Aplicando la fórmula tenemos:
  $$THD = \dfrac{\sqrt{0.2^{2} + 0.1^{2} + 0.05^{2} + 0.03^2}}{1^{2}}$$\
  $$THD = \dfrac{\sqrt{0.0534}}{1}$$\
  $$THD ≈ 0.231$$\
  $$THD ≈ 23.1\%$$

  Al tener THD, utilizamos la fórmula para obtener el factor de potencia con base al resultado obtenido (THD).

  $$f.d.p = DPF \cdot \dfrac{1}{\sqrt{1+0.231^2}}$$\
  $$f.d.p ≈ 0.974 $$


- ### Caso: Carga RC en Serie
  - Fuente AC: $V_o(t)=V_m \cdot sin(ωt)$

  - Frecuencia: $f=60Hz$

  - Carga: $R=100Ω$ y $C = 100μF$

  Si suponemos una magnitud fundamental:
  $I_1 = 0.8A$

  Y las magnitudes de los armónicos superiores:
  
  $I_2=0.5A$,

  $I_3=0.4A$,

  $I_4=0.3A$,

  $I_5=0.2A$

  Entonces:

  $$THD = \dfrac{\sqrt{0.5^{2} + 0.4^{2} + 0.3^{2} + 0.2^2}}{0.8^{2}}$$\
  $$THD = \dfrac{\sqrt{0.54}}{0.8^{2}}$$\
  $$THD ≈ 0.9186$$\
  $$THD ≈ 91.86\%$$

  Conociendo THD, su factor de potencia sería:
  $$f.d.p = DPF \cdot \dfrac{1}{\sqrt{1+0.9186^2}}$$\
  $$f.d.p ≈ 0.736 $$
  

Se observa que n el caso resistivo, aunque hay cierta distorsión, es menor porque la corriente sigue la forma de onda sin interrupciones prolongadas, y n el caso RC, el capacitor se carga rápidamente durante cortos intervalos, lo que provoca corrientes pico muy altas y una forma de onda distorsionada (THD elevado y factor de potencia bajo).

Estos efectos son críticos en aplicaciones reales como fuentes de alimentación conmutadas, donde se requiere corrección activa del factor de potencia.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq
import scipy
from scipy.signal import lti, lsim

# 1. Parámetros de la señal y el sistema
f = 60                    # Frecuencia de red [Hz]
w = 2 * np.pi * f         # Velocidad angular
Vm = 170                  # Voltaje pico (120 Vrms)
T = 1 / f                 # Periodo
t = np.linspace(0, 2*T, 200)  # Tiempo para 2 ciclos
fs = len(t) / (t[-1] - t[0])   # Frecuencia de muestreo

# Señal de entrada: seno ideal de 60 Hz
vin = Vm * np.sin(w * t)

# Rectificación de onda completa
vout = np.abs(vin)

# ------------------------------
# Funciones de análisis
# ------------------------------

def calcular_thd(signal, fs):
    """ Calcula el THD de una señal usando FFT. """
    N = len(signal)
    freqs = fftfreq(N, d=1/fs)
    fft_vals = np.abs(fft(signal))[:N//2]
    freqs = freqs[:N//2]

    fft_vals[0] = 0  # eliminar componente DC si hay

    fundamental_index = np.argmax(fft_vals)
    V1 = fft_vals[fundamental_index]
    pot_total = np.sum(fft_vals**2)
    pot_harm = pot_total - V1**2
    thd = np.sqrt(pot_harm) / V1
    return thd, freqs, fft_vals

def calcular_fp(v, i):
    """ Calcula el factor de potencia: activa / aparente """
    P_activa = scipy.integrate.simpson(v * i, t) / (t[-1] - t[0])
    Vrms = np.sqrt(np.mean(v**2))
    Irms = np.sqrt(np.mean(i**2))
    S_aparente = Vrms * Irms
    return P_activa / S_aparente

# ------------------------------
# i) CARGA NETAMENTE RESISTIVA
# ------------------------------
R = 100  # Ohmios
i_r = vout / R

thd_r, freqs_r, fft_r = calcular_thd(i_r, fs)
fp_r = calcular_fp(vout, i_r)

# ------------------------------
# ii) CARGA RC EN SERIE
# ------------------------------
C = 100e-6  # Faradios
RC = R * C

# Sistema RC en el dominio continuo: H(s) = 1 / (RCs + 1)
num = [1]
den = [RC, 1]
sistema = lti(num, den)

# Simular corriente a través de carga RC con respuesta al voltaje rectificado
_, i_rc, _ = lsim(sistema, U=vout, T=t)

thd_rc, freqs_rc, fft_rc = calcular_thd(i_rc, fs)
fp_rc = calcular_fp(vout, i_rc)

# ------------------------------
# GRAFICAR RESULTADOS
# ------------------------------

plt.figure(figsize=(14, 6))

# Corriente resistiva
plt.subplot(2, 1, 1)
plt.plot(t, i_r, label='Corriente Resistiva')
plt.plot(t, vout / Vm, '--', label='Voltaje Normalizado')
plt.title("Corriente con carga resistiva pura")
plt.ylabel("Amplitud")
plt.grid()
plt.legend()

# Corriente RC
plt.subplot(2, 1, 2)
plt.plot(t, i_rc, label='Corriente RC')
plt.plot(t, vout / Vm, '--', label='Voltaje Normalizado')
plt.title("Corriente con carga RC en serie")
plt.xlabel("Tiempo [s]")
plt.ylabel("Amplitud")
plt.grid()
plt.legend()

plt.tight_layout()
plt.show()

# ------------------------------
# ESPECTROS FFT (Visualización)
# ------------------------------

plt.figure(figsize=(14, 5))

plt.subplot(1, 2, 1)
plt.stem(freqs_r, fft_r, basefmt=" ")
plt.title("FFT de la corriente resistiva")
plt.xlabel("Frecuencia [Hz]")
plt.ylabel("Magnitud")

plt.subplot(1, 2, 2)
plt.stem(freqs_rc, fft_rc, basefmt=" ")
plt.title("FFT de la corriente RC")
plt.xlabel("Frecuencia [Hz]")
plt.ylabel("Magnitud")

plt.tight_layout()
plt.show()

# ------------------------------
# MOSTRAR RESULTADOS
# ------------------------------

print("----- RESULTADOS THD Y FP -----")
print(f"Carga Resistiva:")
print(f"  THD: {thd_r*100:.2f}%")
print(f"  Factor de Potencia: {fp_r:.4f}")

print(f"\nCarga RC en serie:")
print(f"  THD: {thd_rc*100:.2f}%")
print(f"  Factor de Potencia: {fp_rc:.4f}")
```

```python
# Función de transferencia:
s = sym.symbols('s', complex=True)
t, R, L, C = sym.symbols('t R L C', positive=True)
X = sym.Function('X')(s)

Y = 1/(R*C*s + 1) * X
Y
```

```python
# Valores de R y C: (Caso 2)
R_v = 100 #valores de simulacion de R = 100 ohm
C_v =  100e-6 # valores de simulacion de C = 100 uF
Y_RC = Y.subs(R, R_v).subs(C, C_v)
Y_RC
```

```python
# Resolución de la función de transferencia
num = np.array([1]) #coeficientes del numerador de la funcion de transferencia como numpy array
den = np.array([R_v*C_v,1])#coeficientes del denominador de la funcion de transferencia como numpy array
G_n = sig.TransferFunction(num, den) #funcion de transferencia en
print(num)
print(den)
```

```python
# Evaluando el sistema (Entrada contra Salida):

Fo = 60 #frec alimentación
Fs = 30*Fo #frecuencia muestreo
To = 1/Fo #periodo fundamental
Ts = 1/Fs #periodo muestreo

t = np.arange(0, 5*To,Ts) # se simulan 5 peridos de alimentación con un muestreo segun Ts

A = 120 #amplitud entrada

#función de entrada aproximada como onda senoidal rectificada completa
in_o = A*(np.sin(2*np.pi*Fo*t))
rec_c = sig.square(2*np.pi*Fo*t)#función tren de pulsos para simular señal rectificada

#Para rectificador de media onda:
rec_m = 0.5*(sig.square(2*np.pi*Fo*t)+1)
in_ = in_o * rec_c #si se quiere probar con rec media onda multiplicar por rec_m
out = G_n.output(in_, T=t)[1] #evaluar salida ante entrada rectificada mediante aproximacion numérica de scipy

```

```python
#señales de alimentación y tren de pulsos
plt.plot(t,in_o,label='Alimentación')
plt.legend()
plt.show()
```

Observamos una señal sinusoidal de entrada con amplitud pico de aproximadamente 120 V.
- La frecuencia es de 60 Hz (red eléctrica típica).
- La forma de onda es perfectamente sinusoidal, lo cual indica que la fuente de alimentación es ideal sin distorsión.
- Este es el voltaje antes de pasar por el rectificador.

```python
plt.plot(t,rec_c,label='Tren Pulsos')
plt.legend()
plt.show()
```

Aqui se epresenta la salida del rectificador de onda completa, que convierte la señal sinusoidal en una forma de onda pulsante.

- La señal tiene un patrón repetitivo de picos positivos cada semiciclo de la tensión de entrada.
- Los valores negativos de la señal original se eliminan debido al rectificador de onda completa.
- La forma de onda contiene muchos armónicos impares (3ª, 5ª, 7ª, etc.).

```python
plt.plot(t, in_,label='In(t)')#entrada rectificada
plt.plot(t, out,label='Out(t)')#salida del circuito
plt.xlabel('t [s]')
plt.ylabel('out(t)')
plt.show()
```

Se observa la respuesta del circuito RC a la señal rectificada.

- La señal de salida es una versión suavizada de la señal rectificada.
- El capacitor filtra parcialmente las transiciones rápidas de la señal rectificada, pero aún se observan oscilaciones bruscas.
- Las oscilaciones indican que el filtro RC no es suficiente para eliminar completamente los armónicos altos.
- Hubo un caso anterior de C=1μF (por error), y se ve que esta señal parece más suave, lo cual sugiere que el valor de C=100μF mejora la calidad de la señal.

```python
# Espectro

vfre = np.fft.rfftfreq(len(out),1/Fs) #vector de frecuencia en Hz segun tamaño del vector y tiempo de muestreo
Xf = np.fft.rfft(out)

plt.stem(vfre,abs(Xf/len(out)))
plt.xlabel('Frecuencia [Hz]')
plt.ylabel('|X(f)|')
plt.show()
```

- En comparación con el caso anterior (C=1μF ), los armónicos superiores parecen tener magnitudes menores, lo cual es consistente con un capacitor más grande (C=100μF ).
